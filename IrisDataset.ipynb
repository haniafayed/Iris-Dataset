# Histograms for features

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris = load_iris()
colors = ['blue', 'red', 'green', 'magenta']
fig, axes = plt.subplots(4, figsize=(5,15))
for i, color in zip(range(len(iris.feature_names)), colors):
  x = iris.data[:, i]
  axes[i].hist(x, color=color)
  axes[i].set_xlabel(iris.feature_names[i])
  axes[i].set_ylabel("Count")
  axes[i].grid()
plt.show()

```

# Histograms for classes

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris = load_iris()
colors = ['magenta', 'cyan', 'yellow']
fig, axes = plt.subplots(4, figsize=(5,15))
for i in range(len(iris.feature_names)):
  for j, color in zip(range(len(iris.target_names)), colors):
    x = iris.data[iris.target==j, i]
    axes[i].hist(x, label= iris.target_names[j], color = color)
    axes[i].set_xlabel(iris.feature_names[i])
    axes[i].set_ylabel("Count")
    axes[i].grid()
    axes[i].legend(loc='upper right')
plt.show()
```

# Scatter plot for every 2 attributes

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris = load_iris()
colors = ['blue', 'red', 'green']
fig, axes = plt.subplots(6, figsize=(10,25))
ptr = 0
for i in range(0,len(iris.feature_names)):
  for j in range(i+1, len(iris.feature_names)):
    for label, color in zip (range(len(iris.target_names)), colors):
      axes[ptr].scatter(iris.data[iris.target==label, i], iris.data[iris.target==label, j], color=color, label=iris.target_names[label])
      axes[ptr].set_xlabel(iris.feature_names[i])
      axes[ptr].set_ylabel(iris.feature_names[j])
      axes[ptr].legend(loc='upper right')
    ptr = ptr + 1
plt.show()
```

# 3D Scatter Plot

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris = load_iris()
fig = plt.figure(figsize = (15, 15))

ax = fig.add_subplot(2, 2, 1, projection='3d')
ax.scatter3D(iris.data[:,0], iris.data[:,1], iris.data[:,2], color = "purple")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[2])

ax = fig.add_subplot(2, 2, 2, projection='3d')
ax.scatter3D(iris.data[:,0], iris.data[:,1], iris.data[:,3], color = "green")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])
ax.set_zlabel(iris.feature_names[3])

ax = fig.add_subplot(2, 2, 3, projection='3d')
ax.scatter3D(iris.data[:,0], iris.data[:,2], iris.data[:,3], color = "orange")
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[2])
ax.set_zlabel(iris.feature_names[3])

ax = fig.add_subplot(2, 2, 4, projection='3d')
ax.scatter3D(iris.data[:,1], iris.data[:,2], iris.data[:,3], color = "blue")
ax.set_xlabel(iris.feature_names[1])
ax.set_ylabel(iris.feature_names[2])
ax.set_zlabel(iris.feature_names[3])
plt.show()
```

# Correlation Matrix

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
iris = load_iris()
df = pd.DataFrame(iris.data,columns=['sepal length','sepal width','petal length', 'petal width'])
correlation_matrix = df.corr()
sn.heatmap(correlation_matrix, annot=True)
plt.show()
```

**Comment**: Petal length and petal width have a strong positive correlation.
         Sepal length and petal length have a strong positive correlation.
         Sepal length and petal length have a strong positive correlation.
         Strong positive correlation indicates that when one attribute's value increases the other attributes's value increases as well.
         Sepal length and sepal width have almost no correlation and therefore can be said to be independent of each other.

# Boxplots

```python
import seaborn as sn
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize=(15,15))

df = sn.load_dataset('iris')
df.head()

sn.boxplot( y=df["sepal_length"], x=df["species"], ax=axes[0, 0], palette="Set2")
sn.boxplot( y=df["sepal_width"], x=df["species"], ax=axes[0, 1], palette="Paired")
sn.boxplot( y=df["petal_length"], x=df["species"], ax=axes[1, 0], palette="husl")
sn.boxplot( y=df["petal_width"], x=df["species"], ax=axes[1, 1], palette="hls")
plt.show()
```
